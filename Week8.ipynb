```python
import os
import pyfiglet

# السؤال 1
# [span_3](start_span)التحقق من وجود المجلد الرئيسي قبل البدء بعملية الإنشاء[span_3](end_span)
if not os.path.exists("lab_data"):
    # [span_4](start_span)إنشاء المجلد الرئيسي المطلوب في التمرين[span_4](end_span)
    os.makedirs("lab_data")

# السؤال 1
# [span_5](start_span)إنشاء المجلدات الفرعية الثلاثة داخل المجلد الرئيسي[span_5](end_span)
subfolders = ["input", "output", "logs"]
for folder in subfolders:
    path = os.path.join("lab_data", folder)
    # [span_6](start_span)التأكد من عدم إنشاء المجلد إذا كان موجوداً بالفعل[span_6](end_span)
    if not os.path.exists(path):
        os.makedirs(path)

# السؤال 1
# [span_7](start_span)عرض قائمة بجميع المحتويات وتحديد هل هي ملف أم مجلد[span_7](end_span)
print("Contents of lab_data:")
for item in os.listdir("lab_data"):
    item_path = os.path.join("lab_data", item)
    if os.path.isdir(item_path):
        print(f"{item}: Directory")
    else:
        print(f"{item}: File")

# السؤال 2
# [span_8](start_span)تغيير اسم مجلد السجلات إلى اسم مجلد التاريخ[span_8](end_span)
old_logs = os.path.join("lab_data", "logs")
new_history = os.path.join("lab_data", "history")
# [span_9](start_span)التحقق من وجود المجلد القديم وعدم وجود المجلد الجديد لتجنب الخطأ الظاهر في صورتك[span_9](end_span)
if os.path.exists(old_logs) and not os.path.exists(new_history):
    # [span_10](start_span)تنفيذ عملية إعادة تسمية المجلد[span_10](end_span)
    os.rename(old_logs, new_history)

# السؤال 3
# [span_11](start_span)إنشاء ملف نصي جديد داخل مجلد المدخلات وكتابة خمسة أسطر[span_11](end_span)
notes_path = os.path.join("lab_data", "input", "notes.txt")
with open(notes_path, "w") as f:
    # [span_12](start_span)كتابة الأسطر الخمسة المطلوبة برمجياً[span_12](end_span)
    f.write("Line 1: Python programming is powerful\n")
    f.write("Line 2: OS module manages files and folders\n")
    f.write("Line 3: Automation saves time and effort\n")
    f.write("Line 4: This is the fourth line of text\n")
    f.write("Line 5: Final line in the notes file\n")

# السؤال 4
# [span_13](start_span)قراءة الملف وطباعة إجمالي عدد الأسطر الموجودة فيه[span_13](end_span)
with open(notes_path, "r") as f:
    lines = f.readlines()
    # [span_14](start_span)طباعة العدد النهائي للأسطر[span_14](end_span)
    print(f"Total lines in notes.txt: {len(lines)}")

# السؤال 4
# [span_15](start_span)نسخ محتويات الملف إلى ملف جديد داخل مجلد المخرجات[span_15](end_span)
summary_path = os.path.join("lab_data", "output", "summary.txt")
with open(summary_path, "w") as f:
    # [span_16](start_span)كتابة الأسطر المنسوخة في الملف الجديد[span_16](end_span)
    f.writelines(lines)

# السؤال 5، 6، 7، 8
# [span_17](start_span)[span_18](start_span)استخدام المكتبة الخارجية لإنشاء نص رسومي وحفظه في ملف[span_17](end_span)[span_18](end_span)
# [span_19](start_span)[span_20](start_span)تحويل الاسم إلى فن رسومي باستخدام المكتبة المثبتة[span_19](end_span)[span_20](end_span)
my_name_art = pyfiglet.figlet_format("Lily")
art_file = os.path.join("lab_data", "output", "art_output.txt")
with open(art_file, "w") as f:
    # [span_21](start_span)[span_22](start_span)حفظ النتيجة النهائية في ملف نصي مستقل داخل مجلد المخرجات[span_21](end_span)[span_22](end_span)
    f.write(my_name_art)

# السؤال 9
# [span_23](start_span)حذف مجلد التاريخ الفارغ الموجود داخل مجلد المعمل الرابع[span_23](end_span)
# [span_24](start_span)تم استخدام الاسم المطلوب في نص السؤال بدقة لضمان الصحة[span_24](end_span)
cleanup_path = os.path.join("lab4_data", "history")
if os.path.exists(cleanup_path):
    # [span_25](start_span)إزالة المجلد من النظام نهائياً[span_25](end_span)
    os.rmdir(cleanup_path)
```
